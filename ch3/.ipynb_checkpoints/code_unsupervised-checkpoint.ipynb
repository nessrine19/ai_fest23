```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = df_unsuper.values
# Initialize an empty list to store the WCSS values
wcss = []

# Try different values of n_clusters
for n_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  # Append the WCSS to the list

# Plot the WCSS values against the number of clusters
plt.plot(range(1, 11), wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

## Step 3:  Quantifying patient differences

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Convert the cleaned data to numpy array
X = df_unsuper.values
# Apply kmeans
seed = 42

# Initialize the KMeans object with the seed
kmeans = KMeans(n_clusters=3, random_state=seed)#from elbow method
kmeans.fit(X)

# Plot the clustered data
fig, ax = plt.subplots(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')

# Mark the centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=300, alpha=0.7)

plt.show()
```

```python
#without seeds
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = df_unsuper.values
kmeans = KMeans(n_clusters=3)#from elbow method
kmeans.fit(X)

cluster_labels = kmeans.predict(X)

# Create a DataFrame with ID and cluster assignment
result_df = pd.DataFrame({'id': df_unsuper['id'], 'Cluster': cluster_labels})

# Save the DataFrame to a CSV file
result_df.to_csv('cluster.csv', index=False)
# Plot the clustered data
fig, ax = plt.subplots(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')

# Mark the centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=300, alpha=0.7)

plt.show()
```

```python

n_clusters = 3
n_iterations = 10

X = df_unsuper.values
results = pd.DataFrame()

# Perform iterations of K-means clustering
for i in range(n_iterations):
    # Create the KMeans object with the desired number of clusters and random seed
    kmeans = KMeans(n_clusters=n_clusters,random_state=i)
    
    # Fit the K-means model to the data
    kmeans.fit(X)
    
    # Get the labels assigned to each data point
    labels = kmeans.labels_
    
    # Store the clustering results in the DataFrame
    results[f'Iteration {i+1}'] = labels

# Plotting the clustering results
fig, axes = plt.subplots(n_iterations, figsize=(8, 6 * n_iterations))

for i, ax in enumerate(axes):
    ax.scatter(X[:, 0], X[:, 1], c=results[f'Iteration {i+1}'], cmap='viridis')
    ax.set_title(f'Iteration {i+1}')

plt.tight_layout()
plt.show()

# Display the clustering results table
print(results)
```

```python
# Create a DataFrame with the data and cluster labels
X = df_unsuper.values
data_with_clusters = pd.DataFrame(data=X)
data_with_clusters['Cluster'] = labels

# Calculate the mean and mode for each cluster
cluster_statistics = {}

for cluster in np.unique(labels):
    cluster_data = data_with_clusters[data_with_clusters['Cluster'] == cluster].drop('Cluster', axis=1)
    
    # Calculate the mean and mode for each feature in the cluster
    mean_values = cluster_data.mean()
    mode_values = cluster_data.mode().iloc[0]
    
    cluster_statistics[cluster] = {'Mean': mean_values, 'Mode': mode_values}

# Print the cluster statistics
for cluster, statistics in cluster_statistics.items():
    print(f"Cluster {cluster} Statistics:")
    print("Mean:")
    print(statistics['Mean'])
    print("Mode:")
    print(statistics['Mode'])
    print()
```
